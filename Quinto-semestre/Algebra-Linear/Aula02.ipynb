# Matriz

Uma matriz $A$ é uma tabela de $m \times n$ elementos

As letras $a_{ij}$ representam os elementos da matriz. Sendo que $a_{ij}$ é o elemento lovalizado na i-ésima linha e j-ésima coluna da matriz. Por exemplo:

$$
\begin{equation} A =
\begin{pmatrix}
  a_{(0, 0)}       & a_{(0, 1)}   & a_{(0, 2)}   & \cdots  & a_{(0, j)}  \\
  a_{(1, 0)}       & a_{(1, 1)}   & a_{(1, 2)}   & \cdots  & a_{(1, j)}  \\
  \vdots  & \vdots  & \vdots  & \ddots  & \vdots \\
  a_{(0, 0)}       & x^1_n   & x^2_n   & \cdots  & a_{ij}  \\
\end{pmatrix}
\end{equation}
$$

## Tipos de matrizes:

- Matriz retangular =  $m \neq n$
- Matriz quadrada: $m = n$
- matriz linha: Matriz de ordem $1 \times n$
- Matriz coluna: matriz de ordem $n \times 1$

## Diagonais de uma matriz:

- Diagonal principal: Formada por todos os elementos $a_{ij}$ de uma matriz quadrada onde $i = j$
- Diagonal secundária: Formada por todos os elementos $a_{ij}$ onde $i + j = n + 1$

## Matriz identidade:

Matriz quadrada onde todos os elementos da diagonal principal são iguais a 1 e os demais são iguais a 0

## Matriz nula:

Todos os elementos $a_{ij} = 0$

## Matriz transposta:

A transposta de uma matriz $A$, denotada por $A^T$, é obtida trocando-se as linhas por colunas

**Propriedades:**

- $(A+B)^T = A^T + B^T$
- $(\Delta A)^T = \Delta A ^T$
- $(A^T)^T = A$
- $(AB)^T = B^TA^T$

## Matriz normal

Uma matriz quadrada $A$ de odem $n$ é dita normal quando comuda com a sua transposta:

$$A \times A^T = A^T \times A$$

## Igualdade de matrizes

Duas matrizes $A$ e $B$ de ordem $m \times n$ são iguais se, e somente se, $a_{ij} = b_{ij}$

## Traço da matriz

O traço (Tr) de uma matriz quadrada é a soma dos elementos da sua diagonal principal

A multiplicação de uma matriz $A$ e um escalar $k$ (uma matriz $1 \times 1$) pode ser feitam multiplicando todos os elementos da $a_{ij}$ da matriz $A$ pelo escalar $k$

## Multiplicação de matriz
