# Criando a base de dados

Criando os animais com as devidas caracteristicas (Features)

Considerando: 1 - SIM; 0 - NÃO

Features:

1. Tem pelo longo
2. Tem pernas curtas
3. O animal late

Animal = [1, 2, 3]

```
# Criando os animais
#/Quinto-semestre/Inteligencia-Artificial/Aula1.ipynb
#Porcos
animal1 = [0, 1, 0]
animal2 = [0, 1, 1]
animal3 = [1, 1, 0]

#Cachorros
animal4 = [0, 1, 1]
animal5 = [1, 0, 1]
animal6 = [1, 1, 1]

# Classificar os animais

dados = [animal1, animal2, animal3, animal4, animal5, animal6]

# Criando base de classificação dos animais [1 - Porco, 0 - Cachorro]
classes = [1, 1, 1, 0, 0, 0]
```

# Implementação do estimador

Inicialmente (sem entrar em detalhes tecnicos), vamos adotar o modelo [LinearSVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html)


```
# Importando o LinearSVC

from sklearn.svm import LinearSVC

# Selecionando o modelo

model = LinearSVC()

# Entrar com os dados que deseja ensinar e classificar

model.fit(dados, classes)
```

Para testar esse processo, vamos criar um animal.

animal0 -> Vai ter pelos longos, pernas curtas e late.

que animal é esse?

```
animal0 = [0, 0, 1]

# Questionando o algoritimo e modelo utilizado para classificar (prever) que tipo de animal é esse.

pred = model.predict([animal0])

if pred == [0]:
  print("Animal é um cachorro")
elif pred == [1]:
  print("Animal é um porco")
```

Criando varios animais (sem saber qual é a sua classificação)

```
animal10 = [1, 0, 1]
animal11 = [0, 0, 0]
animal12 = [0, 0, 1]

testes = [animal10, animal11, animal12]

pred = model.predict(testes)
print(pred)

#respostas dos animais 10, 11 e 12

testesClasses = [0, 1, 0]

# Comparação da previsão obtida com o resultado correto
pred == testesClasses
```

```
corretos = (pred == testesClasses).sum()
total = len(testes)
taxaAcerto = corretos/total
print(f"Taxa de acerto: {taxaAcerto * 100}%")
```
